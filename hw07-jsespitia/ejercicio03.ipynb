```python
%pylab inline
tamanio = 10
```

```python
cells = np.zeros([tamanio,tamanio],dtype=int)
cellsAnterior = np.zeros([len(cells),len(cells[0])],dtype = int)
```

```python
#Condiciones del array de prueba
cells[3,3] = 1
cells[3,4] = 1
cells[3,5] = 1

cells[6,6] = 1
```

```python
def copiar():
    cellsAnterior = np.zeros([len(cells),len(cells[0])],dtype = int)
    for i in range(0,tamanio):
        for j in range(0,tamanio):
            cellsAnterior[i][j] = cells[i][j]
```

```python
def vivira(x,y):
    cantVecinos = 0
    if(x > 0):
        if(y>0 and cellsAnterior[x-1][y-1]==1):
            cantVecinos+=1
        if(cellsAnterior[x-1][y]==1):
            cantVecinos+=1
        if(y<tamanio-1 and cellsAnterior[x-1][y+1]==1):
            cantVecinos+=1
    if(y > 0):
        if(cellsAnterior[x][y-1]==1):
            cantVecinos+=1
        if(x<tamanio-1 and cellsAnterior[x+1][y-1]==1):
            cantVecinos+=1
    if(y<tamanio-1):
        if(cellsAnterior[x][y+1] ==1):
            cantVecinos+=1
        if(x<tamanio-1 and cellsAnterior[x+1][y+1]==1):
            cantVecinos+=1
    if(x<tamanio-1 and cellsAnterior[x+1][y]):
        cantVecinos+=1
    
    #print(x,y,cellsAnterior)
    #luego de sumar los vecinos, se verifican las reglas
    if(cellsAnterior[x][y] == 1):
        if(cantVecinos != 2 and cantVecinos != 3):
            cells[x][y] = 0
    if(cellsAnterior[x][y] == 0):
        if(cantVecinos == 3):
            cells[x][y] = 1
```

```python

for t in range(0,10):
    copiar()
    fig = plt.figure()
    plt.imshow(cells)
    for x in range(0,tamanio):
        for y in range(0,tamanio):
            vivira(x,y)
```
