```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
### Cargar y visualizar datos

data = np.loadtxt("munich_temperatures_average_with_bad_data.txt")
x = data[:,0]
y = data[:,1]
plt.scatter(x,y)
```

```python
print(y)
```

```python
b = np.where(np.abs(y)<=50)
```

```python
x1 = x[b]
y1 = y[b]
plt.scatter(x1,y1)
```

```python
def f(t,a,b,c,d):
    y = a*np.cos(b*t) + c*np.sin(d*t)
    return y

popt, pcov = curve_fit(f, x1, y1)
```

```python
print("los parámetros de a, b, c y d óptimos son: ")
print(popt)
```

```python
plt.scatter(x1,f(x1,popt[0],popt[1],popt[2], popt[3]))
```
